# Scraping Inderdiscount

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

#---------------------------------------------------------------------------------------------------------
# 1. Schritt: Definition des Drivers und Navigation auf Website
#---------------------------------------------------------------------------------------------------------

# Pfad zum Chromedriver angeben (herunterladen und auf Computer speichern)
driver_path = '/home/student/Documents/chromedriver'

# Erstellen eines Service-Objekts sowie Erstellung eines Treiberobjekts unter Verwendung des Service-Objetks
service = Service(executable_path=driver_path)
driver = webdriver.Chrome(service=service)

# Navigation zur Website und Einstellung des Timers
driver.get('https://www.interdiscount.ch/de/computer-gaming/notebooks/notebooks--c511000?brand=APPLE')
time.sleep(1)

# Zähler für die Anzahl der Klicks auf den "Weiter" Button
click_count = 0

#---------------------------------------------------------------------------------------------------------
# 2. Schritt: Erstellung der Listen um Daten zu speichern und laden der Daten
#---------------------------------------------------------------------------------------------------------

# Erstellung leerer Listen um Daten zu speichern
color_element_1_list = []
color_element_2_list = []
rating_list = []
anzahl_list = []
produktname_list = []
preis_list = []


while True:
    try:
        # Elemente anhand des XPaths mit Selenium finden
        product_links = driver.find_elements(By.XPATH, '//*[@id="TOP_OF_PRODUCTS_OVERVIEW"]/div[4]/div[2]/div[6]/div/div/a')
        num_products = len(product_links)
        
        for i in range(1, num_products + 3):
            # Klicken auf das Produkt, um zur Produktseite zu gelangen
            product_xpath = f'//*[@id="TOP_OF_PRODUCTS_OVERVIEW"]/div[4]/div[2]/div[6]/div[{i}]/div/a'
            product_link = driver.find_elements(By.XPATH, product_xpath)
            
            if not product_link:
                continue
    
            product_link[0].click()
            time.sleep(1)

            # Zusätzliche Informationen sammeln
            try:
                color_elements = driver.find_elements(By.CLASS_NAME, '_2gcnV3')
                color_element_1 = color_elements[0].text if len(color_elements) > 0 else "NA"
                color_element_2 = color_elements[1].text if len(color_elements) > 1 else "NA"
            except:
                color_element_1 = "NA"
                color_element_2 = "NA"

            try:
                rating = driver.find_element(By.CLASS_NAME, '_1X48sk').text
            except:
                rating = "NA"

            try:
                anzahl = driver.find_element(By.CLASS_NAME, '_1r94vm').text
            except:
                anzahl = "NA"

            try:
                produktname = driver.find_element(By.CLASS_NAME, 'uIyEJC._2sh9pz._9YoDdk').text
            except:
                produktname = "NA"

            try:
                preis = driver.find_element(By.CLASS_NAME, 'krweWr.czipba').text
            except:
                preis = "NA"
            
            # Zusatzinformationen den Listen hinzufügen
            color_element_1_list.append(color_element_1)
            color_element_2_list.append(color_element_2)
            rating_list.append(rating)
            anzahl_list.append(anzahl)
            produktname_list.append(produktname)
            preis_list.append(preis)
            
            print("-------------")
            print(color_element_1)
            print(color_element_2)
            print(rating)
            print(anzahl)
            print(produktname)
            print(preis)
            
            # Zurück zur vorherigen Seite navigieren
            driver.back()
            time.sleep(1)

    except:
        break

    # Scrollen zum Anfang der Seite
    driver.execute_script("window.scrollTo(0, 0)")

    # Überprüfen, ob der Weiter-Button vorhanden ist
    next_button = driver.find_elements(By.XPATH, '//*[@id="TOP_OF_PRODUCTS_OVERVIEW"]/div[4]/div[2]/div[7]/ul/li[7]/a/div')
    if not next_button:
        break

    click_count += 1

    if click_count >= 10:
        break
    
    # Auf den Weiter-Button klicken
    next_button[0].click()
    time.sleep(1)

# Browserfenster schließen
driver.quit()



#---------------------------------------------------------------------------------------------------------
# 3. Schritt: Erstellung des Dataframe anhand der Listen und Speicherung als CSV File
#---------------------------------------------------------------------------------------------------------
def source_data():

    # Erstellung Dataframe anand der Listen
    df = pd.DataFrame({
        "color_element_1": color_element_1_list,
        "color_element_2": color_element_2_list,
        "Rating": rating_list,
        "Anzahl": anzahl_list,
        "Produktname": produktname_list,
        "Preis": preis_list
    })

    # Speichern des Dataframes in die Stage 1 CSV-Datei
    df.to_csv('Heini_Moreno_stage1.csv', index=False)

    return df

source_df = source_data()
print(source_df.head())
```
